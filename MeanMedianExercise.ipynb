# Exercise: Mean & Median Customer Spend

Here's some code that will generate some random e-commerce data; just an array of total amount spent per transaction. Select the code block, and hit "play" to execute it:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

Now, find the mean and median of this data. In the code block below, write your code, and see if your result makes sense:

```python
# Calculate mean and median
mean_value = np.mean(incomes)
median_value = np.median(incomes)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"\nDifference: {abs(mean_value - median_value):.2f}")
```

This is pretty much the world's easiest assignment, but we're just trying to get your hands on iPython and writing code with numpy to get you comfortable with it.

Try playing with the code above to generate different distributions of data, or add outliers to it to see their effect.

```python
# Experiment 1: Skewed distribution (positive skew)
print("=" * 50)
print("EXPERIMENT 1: POSITIVE SKEW (Right-skewed)")
print("=" * 50)
skewed_data = np.random.exponential(scale=20, size=10000)
plt.hist(skewed_data, bins=50)
plt.title('Right-Skewed Distribution')
plt.show()
mean_skewed = np.mean(skewed_data)
median_skewed = np.median(skewed_data)
print(f"Mean: {mean_skewed:.2f}")
print(f"Median: {median_skewed:.2f}")
print(f"Difference: {abs(mean_skewed - median_skewed):.2f}")
print(f"Note: Mean > Median because of right tail with large outliers\n")
```

```python
# Experiment 2: Normal distribution WITH extreme outliers
print("\n" + "=" * 50)
print("EXPERIMENT 2: NORMAL DATA WITH EXTREME OUTLIERS")
print("=" * 50)
normal_with_outliers = np.random.normal(100, 20, 10000)
# Add some extreme outliers
outliers = np.array([500, 600, 700])  # Very large values
normal_with_outliers = np.append(normal_with_outliers, outliers)

plt.hist(normal_with_outliers, bins=50)
plt.title('Normal Distribution with Extreme Outliers')
plt.show()

mean_outliers = np.mean(normal_with_outliers)
median_outliers = np.median(normal_with_outliers)
print(f"Mean: {mean_outliers:.2f}")
print(f"Median: {median_outliers:.2f}")
print(f"Difference: {abs(mean_outliers - median_outliers):.2f}")
print(f"Note: Extreme outliers dramatically increase the mean!\n")
```

```python
# Experiment 3: Left-skewed distribution (negative skew)
print("\n" + "=" * 50)
print("EXPERIMENT 3: LEFT-SKEWED (Negative Skew)")
print("=" * 50)
left_skewed = 100 - np.random.exponential(scale=20, size=10000)
plt.hist(left_skewed, bins=50)
plt.title('Left-Skewed Distribution')
plt.show()

mean_left = np.mean(left_skewed)
median_left = np.median(left_skewed)
print(f"Mean: {mean_left:.2f}")
print(f"Median: {median_left:.2f}")
print(f"Difference: {abs(mean_left - median_left):.2f}")
print(f"Note: Mean < Median because of left tail with small outliers\n")
```

```python
# Experiment 4: Robustness of Mean vs Median to outliers
print("\n" + "=" * 50)
print("EXPERIMENT 4: ROBUSTNESS COMPARISON")
print("=" * 50)
print("\nScenario: Small dataset with an extreme outlier")
print("Original data: [10, 12, 15, 18, 20]\n")

original = np.array([10, 12, 15, 18, 20])
print(f"Original Mean: {np.mean(original):.2f}")
print(f"Original Median: {np.median(original):.2f}\n")

print("With outlier: [10, 12, 15, 18, 20, 1000]\n")
with_outlier = np.array([10, 12, 15, 18, 20, 1000])
print(f"With outlier Mean: {np.mean(with_outlier):.2f}")
print(f"With outlier Median: {np.median(with_outlier):.2f}\n")

print(f"Mean increased by: {np.mean(with_outlier) - np.mean(original):.2f}")
print(f"Median increased by: {np.median(with_outlier) - np.median(original):.2f}")
print(f"\nConclusion: Median is MUCH MORE ROBUST to outliers!\nMedian barely changed while mean jumped dramatically.")
```

```python
# Summary: Create a comparison chart of all experiments
print("\n" + "=" * 50)
print("SUMMARY OF ALL EXPERIMENTS")
print("=" * 50)

experiments = [
    ('Normal Distribution', 100.11, 100.22, 0.11),
    ('Right-Skewed', 20.14, 13.89, 6.25),
    ('Left-Skewed', 80.12, 86.38, 6.26),
    ('Small + Outlier', 179.17, 16.50, 162.67)
]

import pandas as pd
df = pd.DataFrame(experiments, columns=['Distribution Type', 'Mean', 'Median', 'Difference'])
print("\n", df.to_string(index=False))

print("\n" + "=" * 50)
print("KEY INSIGHTS:")
print("=" * 50)
print("\n1. NORMAL DISTRIBUTION:")
print("   - Mean and Median are nearly equal")
print("   - Symmetric distribution, no skew")

print("\n2. RIGHT-SKEWED (Positive Skew):")
print("   - Mean > Median")
print("   - Long tail on the right pulls mean higher")

print("\n3. LEFT-SKEWED (Negative Skew):")
print("   - Mean < Median")
print("   - Long tail on the left pulls mean lower")

print("\n4. ROBUSTNESS TO OUTLIERS:")
print("   - Median is MUCH more resistant to outliers")
print("   - Single extreme value barely affects median")
print("   - Mean is highly influenced by outliers")
```
